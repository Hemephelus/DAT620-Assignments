```python
import pandas as pd

df = pd.read_csv('ETH-USD.csv')
```

```python
df.head()
```

```python
df_train = df[['date','average_usd_size']]
df_train = df_train.rename(columns={"date": "ds", "average_usd_size": "y"})
df_train = df_train.dropna()
```

```python
df_train
```

```python
from neuralprophet import NeuralProphet

m = NeuralProphet()
m.fit(df_train, freq="D")
```

```python
future = m.make_future_dataframe(df_train, periods=30)
forecast = m.predict(future)
forecast = forecast.rename(columns={"ds": "Date", "yhat1": "Close"})
forecast
```

```python
import pickle
with open('neuralProphet.pkl', "wb") as f:
    pickle.dump(m, f)
```
